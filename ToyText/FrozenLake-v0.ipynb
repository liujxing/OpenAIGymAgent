### Structure of Algorithm for FrozenLake

**Algorithm Overview**

The algorithm contains the following components:
1. learning rate scheduler, responsible for decreasing the learning rate with more episodes
2. exploration probability scheduler, responsible for decreasing the exploration probability with more episodes
3. q-function updater, responsible for update the q-function

The algorithm proceeds in the following matter:
1. At the start of episode, calculate the learning rate and exploration probability given number of episodes.
2. With exploration probability, either randomly explore the action space or take the current optimum action.
3. Record the reward, and use the following algorithm to update the q-function
$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha_t [r_{t+1} + \gamma \space max_a Q(s_{t+1}, a) - Q(s_t, a_t)]$$
